# Regression Modeling in TensorFlow

```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model
tf.__version__
```

## Creating Data


```python
import numpy as np
import matplotlib.pyplot as plt

# Creating Features
x = np.array([-7., -4., -1., 2., 5., 8., 11., 14.])

# Creating Labels
y = np.array([3., 6., 9., 12., 15., 18., 21., 24.])

plt.scatter(x, y)
```

```python
y == x + 10
```

## Input and Output Shapes

```python
# Demo Tensor for the Housing Price Preddiction
house_info = tf.constant(["bedroom", "bathroom", "garage"])
house_price = tf.constant([939700])
house_info, house_price
```

```python
# Doesn't work we don't want to see all the data at once to see a single output
input_shape = x.shape
output_shape = y.shape

input_shape, output_shape
```

```python
# We are only using scalars here or a zero-th dimension tensor
X_train = tf.constant(x, dtype=tf.float32)
Y_train = tf.constant(y, dtype=tf.float32)

X_train,Y_train
```

## Steps in Modelling with tensorflow
1. Create a model
  * define input and output
  * hidden layers
2. Complie the Model
  * define loss function
3. Fit the Model

```python
tf.random.set_seed(42)

# Create a model from Sequential API
model = tf.keras.Sequential([
        tf.keras.Input(shape=(1,)),
        tf.keras.layers.Dense(1)
])

# Compile the Model
model.compile(loss=tf.keras.losses.MAE, # Mean Absoutle Error
              optimizer=tf.keras.optimizers.SGD(), # Stochastic Gradient Decent
              metrics=['mae'])

# Fit the Model
model.fit(X_train, Y_train, epochs=5)
```

```python
# Check out x and y
x ,y
```

```python
# Try and Make a prediction
y_pred = model.predict([17.])
y_pred
```

```python
bigger_x_train = bigger_x[:40]
bigger_y_train = bigger_y[:40]

bigger_x_test = bigger_x[40:]
bigger_y_test = bigger_y[40:]

bigger_x_train, bigger_x_test, bigger_y_train, bigger_y_test
```

```python
def plotting_func(training_data=bigger_x_train, 
                  training_labels=bigger_y_train, 
                  test_data=bigger_x_test, 
                  test_labels=bigger_y_test, 
                  predictions=y_pred):
  plt.figure(figsize=(10,7))
  # Plot training data in blue
  plt.scatter(training_data, training_labels, c="b", label="Training Data")
  #Plot test data in green
  plt.scatter(test_data, test_labels, c="g", label="Testing Data")
  y_pred = bigger_model.predict(bigger_x_test)
  plt.scatter(test_data, predictions, c="r", label="Model's Predictions")
  plt.legend()
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

ct = make_column_transformer(
    (MinMaxScaler(), ["age", "bmi", "children"]),
    (OneHotEncoder(handle_unknown="ignore"), ["sex", "smoker", "region"])
)

# create new x and y
X = insurence.drop("charges", axis=1)
Y = insurence["charges"]

x_train_2, x_test_2, y_train_2, y_test_2 = train_test_split(X, Y, test_size=0.2, random_state=42)

#Fit the columns
ct.fit(x_train_2)

#transform the traing and test data
x_train_normal = ct.transform(x_train_2)
x_test_normal = ct.transform(x_test_2)
```
